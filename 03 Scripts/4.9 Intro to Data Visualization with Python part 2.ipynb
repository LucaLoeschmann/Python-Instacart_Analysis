# 7. connection between age and spending power

```python
scatter_age_income = sns.scatterplot(x = 'age', y = 'income', data = orders_products_all)

scatter_age_income.set(xlabel = 'Customer Age', ylabel = 'Income', title = 'Income by Age')
```

Findings: There is a relationship between age and income. There is a significant increase at the age of 40. There doesn't seem to be a noticable difference for the ages between 40 and 80.
