# Step 1: Create "price_label" and "busiest day" columns

```python
# Subset of the first million rows
df = df_ords_prods_merge[:1000000]
```

```python
# Define function of the price labels
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df.shape
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
df['prices'].max()
```

```python
# Apply price range on df subset with loc() method

df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

```python
 # Use for-loop for "busiest day" (0 = Saturday), "least busy" (4 = Wednesday), or "regularly busy" (other days of the week)

result = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

# Step 2: Change Busiest day” to become “Busiest days” 

```python
# Create column “busiest_days" from "orders_day_of_week" 
# busiest days = 0, 1
# slowest days = 4, 3
# regularly busy = all other days

result_2 = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0 or value == 1:
    result_2.append("Busiest days")
  elif value == 4 or value == 3:
    result_2.append("Slowest days")
  else:
    result_2.append("Regularly busy")
```

# Step 4: Create new column "busiest_period_of_day" to identify time periods

```python
# Check value counts in "order_hour_of_day" column.

df_ords_prods_merge['hour_of_day_ordered'].value_counts()
```

The periods of time are divided into three parts, namely "Most orders", "Average orders", and "Fewest orders".The hours should be divided into three equal groups. The time from 9  to 16  constitutes the "Most orders," the time from 23 to 6  forms the "Fewest orders", and the remaining times,7 to 8 and 17 to 22 make up the "Average orders".

```python
# Create for-loop if statement 
result_3 = []

for value in df_ords_prods_merge["hour_of_day_ordered"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    result_3.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    result_3.append("Fewest orders")
  else:
    result_3.append("Average orders")
```

```python
result_3
```

```python
# Create new column "busiest_period_of_day" in ords_prods_merge.

df_ords_prods_merge['busiest_period_of_day'] = result_3
```
